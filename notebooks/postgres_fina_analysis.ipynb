```python
import pandas as pd
from sqlalchemy import create_engine

# Define your SQLAlchemy engine
engine = create_engine("postgresql+psycopg2://postgres:jack@localhost:5432/telecom")

# Load data into a DataFrame using SQLAlchemy
try:
    df = pd.read_sql("SELECT * FROM xdr_data;", engine)
    print("Data loaded successfully! For agregating of data")

    # Drop rows where 'Bearer Id' is null
    df.dropna(subset=['Bearer Id'], inplace=True)

    # Aggregate the required metrics
    aggregated_data = df.groupby('MSISDN/Number').agg(
        Total_Session_Duration=('Dur. (ms)', 'sum'),
        Total_DL_Bytes=('Total DL (Bytes)', 'sum'),
        Total_UL_Bytes=('Total UL (Bytes)', 'sum')
    ).reset_index()

    # Calculate total volume separately
    aggregated_data['Total_Volume_Bytes'] = (
        df.groupby('MSISDN/Number')['HTTP DL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['HTTP UL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Social Media DL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Social Media UL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Google DL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Google UL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Email DL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Email UL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Youtube DL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Youtube UL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Netflix DL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Netflix UL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Gaming DL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Gaming UL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Other DL (Bytes)'].sum() +
        df.groupby('MSISDN/Number')['Other UL (Bytes)'].sum()
    )

    # Print the aggregated results
    print(aggregated_data)

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Function to identify outliers using Z-score method
def remove_outliers(data, threshold=3):
    z_scores = np.abs((data - data.mean()) / data.std())
    return data[z_scores < threshold]

# Apply the function to relevant columns
for col in df.select_dtypes(include=[np.number]).columns:
    df[col] = remove_outliers(df[col])
```

```python
from sklearn.cluster import KMeans

# Define your SQLAlchemy engine
engine = create_engine("postgresql+psycopg2://postgres:jack@localhost:5432/telecom")

# Load data into a DataFrame using SQLAlchemy
try:
    df = pd.read_sql("SELECT * FROM xdr_data;", engine)
    print("Data loaded successfully!")

    # Drop rows where 'Bearer Id' is null
    df.dropna(subset=['Bearer Id'], inplace=True)

    # Calculate total traffic (DL + UL)
    df['Total_Traffic'] = df['Total DL (Bytes)'] + df['Total UL (Bytes)']

    # Aggregate engagement metrics per customer (MSISDN)
    engagement_metrics = df.groupby('MSISDN/Number').agg(
        Session_Frequency=('Dur. (ms)', 'count'),
        Total_Session_Duration=('Dur. (ms)', 'sum'),
        Total_Traffic=('Total_Traffic', 'sum')
    ).reset_index()

    # Normalize the engagement metrics
    scaler = StandardScaler()
    normalized_metrics = scaler.fit_transform(engagement_metrics[['Session_Frequency', 'Total_Session_Duration', 'Total_Traffic']])
    
    # Run K-means clustering with k=3
    kmeans = KMeans(n_clusters=3, random_state=0)
    engagement_metrics['Cluster'] = kmeans.fit_predict(normalized_metrics)

    # Compute statistics for each cluster
    cluster_stats = engagement_metrics.groupby('Cluster').agg(
        Min_Session_Frequency=('Session_Frequency', 'min'),
        Max_Session_Frequency=('Session_Frequency', 'max'),
        Avg_Session_Frequency=('Session_Frequency', 'mean'),
        Total_Session_Frequency=('Session_Frequency', 'sum'),
        Min_Total_Session_Duration=('Total_Session_Duration', 'min'),
        Max_Total_Session_Duration=('Total_Session_Duration', 'max'),
        Avg_Total_Session_Duration=('Total_Session_Duration', 'mean'),
        Total_Session_Duration=('Total_Session_Duration', 'sum'),
        Min_Total_Traffic=('Total_Traffic', 'min'),
        Max_Total_Traffic=('Total_Traffic', 'max'),
        Avg_Total_Traffic=('Total_Traffic', 'mean'),
        Total_Traffic=('Total_Traffic', 'sum')
    ).reset_index()

    print("Cluster Statistics:\n", cluster_stats)

    # Aggregate user total traffic per application
    application_traffic = df.groupby('MSISDN/Number').agg(
        Total_Application_Traffic=('Total_Traffic', 'sum')
    ).reset_index()

    # Get the top 10 most engaged users per application
    top_users = application_traffic.nlargest(10, 'Total_Application_Traffic')
    print("Top 10 Engaged Users:\n", top_users)

    # Plot the top 3 most used applications
    application_columns = ['HTTP DL (Bytes)', 'Social Media DL (Bytes)', 'Google DL (Bytes)']  # Adjust as necessary
    application_totals = df[application_columns].sum()

    plt.figure(figsize=(10, 6))
    application_totals.plot(kind='bar', color=['blue', 'orange', 'green'])
    plt.title('Total Traffic by Application')
    plt.ylabel('Total Bytes')
    plt.xlabel('Applications')
    plt.xticks(rotation=45)
    plt.show()

    # Elbow method to determine optimal k
    inertia = []
    K = range(1, 11)
    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=0)
        kmeans.fit(normalized_metrics)
        inertia.append(kmeans.inertia_)

    plt.figure(figsize=(10, 6))
    plt.plot(K, inertia, 'bo-')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia')
    plt.title('Elbow Method for Optimal k')
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```

```python

from scipy import stats


# Load data into a DataFrame using SQLAlchemy
try:
    df = pd.read_sql("SELECT * FROM xdr_data;", engine)
    print("Data loaded successfully!")

    # Check for missing values
    missing_values = df.isnull().sum()
    print("Missing Values:\n", missing_values)
    
    df.dropna(subset=['Bearer Id'], inplace=True)
    df.dropna(subset=['MSISDN/Number'], inplace=True)
    # Handle missing values
    for col in df.columns:
        if df[col].dtype == 'object':
            # Replace missing values with mode for categorical variables
            mode_value = df[col].mode()[0]
            df[col].fillna(mode_value, inplace=True)
        else:
            # Replace missing values with mean for numerical variables
            mean_value = df[col].mean()
            df[col].fillna(mean_value, inplace=True)

    # Handle outliers using Z-score method for numerical variables only
    for col in df.select_dtypes(include=[np.number]).columns:
        z_scores = np.abs(stats.zscore(df[col]))
        df = df[z_scores < 3]  # Filter out outliers

    # Aggregate metrics per customer (MSISDN)
    user_experience_metrics = df.groupby('MSISDN/Number').agg(
        Average_TCP_Retransmissions=('TCP DL Retrans. Vol (Bytes)', 'mean'),
        Average_RTT=('Avg RTT DL (ms)', 'mean'),  # Example for DL RTT
        Handset_Type=('IMEI', lambda x: x.mode()[0]),  # Mode for the most common handset
        Average_Throughput=('Avg Bearer TP DL (kbps)', 'mean')  # Example for DL Throughput
    ).reset_index()

    print("User Experience Metrics:\n", user_experience_metrics)

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

try:
    df = pd.read_sql("SELECT * FROM xdr_data;", engine)
    print("Data loaded successfully!")

    # Handle missing values
    for col in df.columns:
        if df[col].dtype == 'object':
            mode_value = df[col].mode()[0]
            df[col].fillna(mode_value, inplace=True)
        else:
            mean_value = df[col].mean()
            df[col].fillna(mean_value, inplace=True)

    # Handle outliers using Z-score method
    for col in df.select_dtypes(include=[np.number]).columns:
        z_scores = np.abs(stats.zscore(df[col]))
        df = df[z_scores < 3]  # Filter out outliers

    # Top, Bottom, and Most Frequent Values
    tcp_values = df['TCP DL Retrans. Vol (Bytes)']
    rtt_values = df['Avg RTT DL (ms)']
    throughput_values = df['Avg Bearer TP DL (kbps)']

    top_tcp = tcp_values.nlargest(10)
    bottom_tcp = tcp_values.nsmallest(10)
    most_frequent_tcp = tcp_values.value_counts().nlargest(10)

    top_rtt = rtt_values.nlargest(10)
    bottom_rtt = rtt_values.nsmallest(10)
    most_frequent_rtt = rtt_values.value_counts().nlargest(10)

    top_throughput = throughput_values.nlargest(10)
    bottom_throughput = throughput_values.nsmallest(10)
    most_frequent_throughput = throughput_values.value_counts().nlargest(10)

    print("Top TCP Values:\n", top_tcp)
    print("\nBottom TCP Values:\n", bottom_tcp)
    print("\nMost Frequent TCP Values:\n", most_frequent_tcp)

    print("\nTop RTT Values:\n", top_rtt)
    print("\nBottom RTT Values:\n", bottom_rtt)
    print("\nMost Frequent RTT Values:\n", most_frequent_rtt)

    print("\nTop Throughput Values:\n", top_throughput)
    print("\nBottom Throughput Values:\n", bottom_throughput)
    print("\nMost Frequent Throughput Values:\n", most_frequent_throughput)

    # Distribution of Average Throughput per Handset Type
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Handset Type', y='Avg Bearer TP DL (kbps)', data=df)
    plt.title('Distribution of Average Throughput per Handset Type')
    plt.xticks(rotation=45)
    plt.show()

    # Average TCP Retransmission per Handset Type
    avg_tcp_per_handset = df.groupby('Handset Type')['TCP DL Retrans. Vol (Bytes)'].mean().reset_index()
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Handset Type', y='TCP DL Retrans. Vol (Bytes)', data=avg_tcp_per_handset)
    plt.title('Average TCP Retransmission per Handset Type')
    plt.xticks(rotation=45)
    plt.show()

    # K-means clustering
    experience_metrics = df.groupby('MSISDN/Number').agg(
        Average_TCP_Retransmissions=('TCP DL Retrans. Vol (Bytes)', 'mean'),
        Average_RTT=('Avg RTT DL (ms)', 'mean'),
        Average_Throughput=('Avg Bearer TP DL (kbps)', 'mean')
    ).reset_index()

    # Normalize the metrics for clustering
    scaler = StandardScaler()
    normalized_metrics = scaler.fit_transform(experience_metrics[['Average_TCP_Retransmissions', 'Average_RTT', 'Average_Throughput']])

    # Run K-means clustering with k=3
    kmeans = KMeans(n_clusters=3, random_state=0)
    experience_metrics['Cluster'] = kmeans.fit_predict(normalized_metrics)

    # Describe each cluster
    cluster_description = experience_metrics.groupby('Cluster').agg(
        Count=('MSISDN/Number', 'count'),
        Mean_TCP_Retransmissions=('Average_TCP_Retransmissions', 'mean'),
        Mean_RTT=('Average_RTT', 'mean'),
        Mean_Throughput=('Average_Throughput', 'mean')
    ).reset_index()

    print("Cluster Descriptions:\n", cluster_description)

except Exception as e:
    print(f"An error occurred: {e}")
```
